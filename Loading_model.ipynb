```python
import tensorflow as tf
import keras
import os
import matplotlib.pyplot as plt
import cv2
from tensorflow.keras.preprocessing import image
import numpy as np
import tensorflowjs as tfjs
from keras.models import model_from_json
```

```python
from tensorflow.keras.models import load_model
new_model = load_model('/content/drive/MyDrive/CNN Model.h5')
```

```python
new_model.summary()
```

```python
img = image.load_img("/content/drive/MyDrive/CNN Model/images/testing/old_back/old_back_1.jpg",
                    target_size=(200,200))
plt.imshow(img)
plt.show()
    
x=image.img_to_array(img)
x=np.expand_dims(x,axis=0)
images = np.vstack([x])
val= val= np.argmax(new_model.predict(images))
if val == 0:
    print("new back")
elif val == 1:
    print("new front")
elif val == 2:
    print("old back")
else :
    print("old front")
```
